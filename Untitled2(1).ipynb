```python
df_ca_by_city = (df_total['price'] * df_total['quantity']).groupby(df_total['city']).sum().reset_index(name='ca')
df_ca_by_city
```

```python
sizes = df_ca_by_city['ca']
explode = [0.1 if s == max(sizes) else 0 for s in sizes]

plt.pie(
    df_ca_by_city['ca'],
    labels=df_ca_by_city['city'],
    autopct=lambda p: f'{p:.1f}%\n({int(p*sum(sizes)/100)})',
    shadow=True,
    explode=explode,
    
)
plt.title('CA par ville')
plt.show()
```

```python
#2- Les villes avec le plus de clients actifs ?
nb_actif = df_total['customer_id'].groupby(df_total['city']).nunique().reset_index(name='nb')
nb_actif
```

```python
sizes = nb_actif['nb']
explode = [0.1 if s == max(sizes) else 0 for s in sizes]

plt.pie(
    nb_actif['nb'],
    labels=df_ca_by_city['city'],
    autopct=lambda p: f'{p:.1f}%\n({int(p*sum(sizes)/100)})',
    shadow=True,
    explode=explode,
    
)
plt.title('Ville par nb de client actif')
plt.show()
```

```python
#3-  Les villes avec le plus de commandes ? 
nb_commande = df_total['city'].value_counts().reset_index(name='commande')
nb_commande
```

```python
sizes = nb_commande['commande']
explode = [0.1 if s == max(sizes) else 0 for s in sizes]

plt.pie(
    nb_commande['commande'],
    labels=df_ca_by_city['city'],
    shadow=True,
    explode=explode,
    autopct=lambda p: f'{p:.1f}%\n({int(p*sum(sizes)/100)})',

    
)
plt.title('Ville avec le plus de commandes')
plt.show()
```
